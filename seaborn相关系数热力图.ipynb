```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = r"D:\下载\machine_learning\机器学习项目\加州房价预测(Hands-On Machine Learning with Scikit-Learn and TensorFlow)\housing.csv"
data= pd.read_csv(path)
```

```python
data.head()
```

```python
data_one_hot_processed = pd.get_dummies(data, columns=['ocean_proximity'])
```

```python
data_one_hot_processed.head()
```

### 原始相关系数热力图

```python
plt.figure(figsize=[10,8])
sns.heatmap(data_one_hot_processed.corr())
plt.show()
```

### 添加注释

```python
plt.figure(figsize=[10,8])
sns.heatmap(data_one_hot_processed.corr(), annot=True)
plt.show()
```

### 保留小数点后面两位

```python
plt.figure(figsize=[10,8])
sns.heatmap(data_one_hot_processed.corr(), annot=True, fmt='.2f')
plt.show()
```

### 添加自定义的color map

```python
plt.figure(figsize=[10,8])
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(data_one_hot_processed.corr(), cmap=cmap, annot=True, fmt='.2f')
plt.show()
```

### 只保留半边

```python
plt.figure(figsize=[10,8])
mask = np.zeros_like(data_one_hot_processed.corr(), dtype=np.bool)  
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(data_one_hot_processed.corr(), mask=mask, cmap=cmap, annot=True, fmt='.2f')
plt.show()
```
